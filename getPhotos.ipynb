```python
#The purpose of this notebook is to make Google Places API calls in order to receive a photo reference, then another API call-
#that will give the actual photo from the reference. It will then convert the photo reference to a .jpg file and save the images
#locally. The images are intended to be used in the tooltips for our markers in the final visualizations.


#import dependencies
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import os
import json
import gmaps
from pprint import pprint
import matplotlib.image as mpimg
from io import StringIO
from PIL import Image
import urllib
import base64
gkey = 'AIzaSyDyMt_HQPR4rNeRsH_D1Is5lbwayhH607A'
```

```python
#import csv
top_europe = pd.read_csv('europesctop.csv')
top_europe
```

```python
#first API call. This will retrieve a photo reference to use for the Google Places Photo API and append to a list of references.
#If there is no photo reference, string 'N/A' will be appended to the list instead
params = {}
params['key'] = gkey
params['rankby'] = 'distance'
base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
references = []
for index, row in top_europe.iterrows():
    hotel_lat = row['lat']
    hotel_lng = row['lng']
    params['location'] = f"{hotel_lat}, {hotel_lng}"
    
    google_response = requests.get(base_url, params = params)
    google_json = google_response.json()
    try:
        pprint(google_json['results'])
        references.append(google_json['results'][0]['photos'][0]['photo_reference'])
    except:
        references.append('N/A')
#     json_string = json.dumps(google_json)
#     json_dict = json.loads(json_string)
    
    
    #print(json.dumps(google_response, indent = 4, sort_keys = True))
    
    
#photoreference = google_json['results'][0]['photos']['photo_reference']
```
